## 1. Calculate the Total Bill
A restaurant bill is ₹450. GST is 18%. Calculate the final bill amount.

```python
# Write your solution here
bill=450
gst=0.18*bill
final_bill=bill+gst
print(final_bill)
```

## 4. Movie Ticket Cost Calculator
Movie ticket = 180, popcorn = 120. Calculate total for 2 tickets + 1 popcorn and 3 tickets + 0 popcorn.

```python
# Write your solution here
movie_ticket=180
popcorn=120
bill_one=2*movie_ticket+popcorn
bill_two=3*movie_ticket+0*popcorn
print(bill_one)
print(bill_two)
```
